```python
Python Logical Operators
Logical operators are used to combine conditional statements:

Operator	Description	Example	Try it
and 	Returns True if both statements are true	
x < 5 and  x < 10	
or	Returns True if one of the statements is true	
x < 5 or x < 4	
not	Reverse the result, returns False if the result is true	not
(x < 5 and x < 10)
```

```python
#Logical operator
a=5
b=10
c=5
ans= (a<b) and (b>c)
print(ans)
```

```python
Python Identity Operators
Identity operators are used to compare the objects,
not if they are equal, but if they are actually the same
object, with the same memory location:
Operator	Description	Example	
is 	        Returns True if both variables are the same object	
      x is y	
is not	     Returns True if both variables are not the same object	
x is not y
```

```python
a=5
b=5
print(id(a))
print(id(b))
print(a is b)
```

```python
a=5
b=15
c=5
print(id(a))
print(id(b))
print(id(c))
print(a is c)
```

```python
a=5
b=12
print(id(a))
print(id(b))

print(a is not b)
```

```python
#example
x = ["apple", "banana"]  #list
y = ["apple", "banana"]
z = x
print(id(x))
print(id(y))
print(id(z))
print(x is z)# returns True because z is the same object as x
print(x is y)# returns False because x is not the same object as y, even if they have the same content
#print(x == y)# to demonstrate the difference betweeen "is" and "==": this comparison returns True because x is equal to y
```

```python
#Python Membership Operators
Membership operators are used
to test if a sequence is presented in an object:
    
Operator	Description	Example	
in 	
Returns True if a sequence with the specified value is present in the object	
x in y	
not in	Returns True if a sequence with the specified value is not present in the object	
x not in y
```

```python
#Examples
x = ["apple", "banana"]

print("banana" in x)
# returns True because a sequence with the value "banana" is in the list
```

```python
x = ["apple", "banana"]

print("mango" not in x)
# returns True because a sequence with the value "mango" is not in the list
```
