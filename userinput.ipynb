```python
#In Python, you can take user input using the built-in input() function.
#It always returns the input as a string, so you may need to convert it to other types (like int or float) if required.

# Taking a simple string input
name = input("Enter your name: ")
print("Hello,", name)
print(type(name))
rollno =int( input("enter stud rollno"))
print("stud rollno is: ",rollno)
print(type(rollno))
per =float( input("enter per"))
print("per is: ",per)
print(type(per))
result=bool(input("enter res for pass or fail(t/f)"))
print("result is: ",result)
print(type(result))
```
