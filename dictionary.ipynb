```python
Dictionary 
a dictionary is a built‑in data structure that stores data in key–value pairs. and can be referred to by using the key name.
It is an ordered collection of data of different  data types.
can be change. created using curly brackets.
Dictionary items are ordered, changeable, and does not allow duplicates.  
```

```python
#Python - Remove Dictionary Items
Removing Items
There are several methods to remove items from a dictionary:
```

```python
#The pop() method removes the item with the specified key name
thisdict =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.pop("model")
print(thisdict)
```

```python
#The popitem() method removes the last inserted item (in versions before 3.7,
#a random item is removed instead):
thisdict =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.popitem()
print(thisdict)
```

```python
#The del keyword removes the item with the specified key name:
thisdict =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
del thisdict["model"]
print(thisdict)
```

```python
#The clear() method empties the dictionary:
thisdict =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
thisdict.clear()
print(thisdict) 
```

```python
# Step 1: Create an empty dictionary to store student names and grades
students = {}

# Step 2: Menu options for user
while True:
    print("\nStudent Grades Tracker")
    print("1. Add Student")
    print("2. Update Grade")
    print("3. Delete Student")
    print("4. Display All Students")
    print("5. Calculate Average Grade")
    print("6. Exit")

    choice = input("Enter your choice (1-6): ")

    if choice == '1':
        name = input("Enter student name: ")
        grade = float(input("Enter student grade: "))
        students[name] = grade
        print(f"{name} added successfully.")

    elif choice == '2':
        name = input("Enter student name to update: ")
        if name in students:
            grade = float(input("Enter new grade: "))
            students[name] = grade
            print(f"{name}'s grade updated.")
        else:
            print("Student not found.")

    elif choice == '3':
        name = input("Enter student name to delete: ")
        if name in students:
            del students[name]
            print(f"{name} deleted.")
        else:
            print("Student not found.")

    elif choice == '4':
        print("\nStudent Grades:")
        for name, grade in students.items():
            print(f"{name}: {grade}")

    elif choice == '5':
        if students:
            average = sum(students.values()) / len(students)
            print(f"Average Grade: {average:.2f}")
        else:
            print("No student data available.")

    elif choice == '6':
        print("Exiting program.")
        break    

    else:
        print("Invalid choice. Please try again.")
```
