```python
Array Module in Python
The array module in Python is a built-in module that provides an efficient way to create and manipulate arrays of basic values such as 
characters, integers, and floating-point numbers. Arrays in Python are similar to lists but are more memory-efficient and can only store 
elements of the same type.

Creating an Array
To create an array, you need to import the array module and use the array class. The array class requires a type code to specify the type of 
elements it will hold. Here is an example of creating an array of integers:
```

```python
import array

# Create an array of integers
numbers = array.array('i', [1, 3, 5, 7])
print(numbers)
```

```python
Type Codes
The type code is a single character that specifies the type of elements in the array. Some common type codes are:
'i': signed integer
'f': float
'd': double
You can get a string representing all the type codes using the typecodes attribute:
```

```python
import array
print(array.typecodes)
```

```python
import array as arr
a = arr.array('i', [1, 2, 3])

# accessing First Araay
print(a[0])

# adding element to array
a.append(5)
print(a)
```

```python
import array as arr
a = arr.array('i', [1, 2, 3])

for i in range(0, 3):
    print(a[i], end=" ")
```

```python
import array as arr
a = arr.array('u', ['a','b','c'])

# accessing First Araay
print(a[0])

# adding element to array
a.append('d')
print(a)
```

```python
array('u') was used for Unicode characters, but now Python recommends using normal strings or other containers instead.
    
```

```python
# using list rather than 'u'
a = ['a', 'b', 'c']
# accessing first element
print(a[0])

# adding element
a.append('d')
print(a)
```

```python
# If you want strings of any length, use a normal Python list instead
a = ["abc", "bio"]

print(a[0])   # abc
print(a)      # ['abc', 'bio']
```

```python
import array as arr
a = arr.array('i', [1, 2, 3])
print(*a)

a.insert(1, 4)  # Insert 4 at index 1
print(*a)
```

```python
Array vs List
The main difference between arrays and lists in Python is that arrays can only store elements of the same type, while lists can store elements
of different types. Arrays are more memory-efficient and provide better performance for numerical operations.
```

```python
import array

# Array of integers
numbers = array.array('i', [1, 3, 5, 7, 1, 2])
print(numbers)

# List of mixed types
numbers2 = [2, "lemon", 3.4]
print(numbers2)
```
