```python
#Python program to drop all digits from a string.

digits = [str(x) for x in range(10)]
mystr = 'He12llo, Py00th55on!'
chars = []
for x in mystr:
    if x not in digits:
        chars.append(x)
newstr = ''.join(chars)
print (newstr)
```

```python
# reverse the string 
s="hello"
n=len(s)
i=0
print("Forward direction")
while i<n:
    print(s[i],end='')
    i +=1
print("\n Backward direction")
i=-1
while i>=-n:
    print(s[i],end=' ')
    i=i-1
```

```python
#altrenative ways
s="hello"
print("Forward direction")
for i in s:
    print(i,end=' ')

print("Forward direction")
for i in s[::]:
    print(i,end=' ')
    
print("Backward direction")
for i in s[::-1]:
    print(i,end=' ')
```

```python
#Python program to find number of words in a string

# Example usage
input_string = "Python is  powerful programming language"

words=input_string.split()
count= len(words)
print("Input string:", input_string)
print("Number of words:", count)
```

```python
# using fun
def count_words(s):
    words = s.split()
    return len(words)

# Example usage
input_string = "Python is a powerful programming language"
word_count = count_words(input_string)
print("Input string:", input_string)
print("Number of words:", word_count)
```

```python
#Program to display all positions of substring in a given main string

s=input("Enter main string:")
subs=input("Enter sub string:")

flag=False
pos=-1

n=len(s)
while True:
    pos=s.find(subs,pos+1,n)
    if pos==-1:
        break
    print("Found at position",pos)

    flag=True
if flag==False:
    print("Not Found")
```
