```python
#Change Tuple Values
fruit_tuple=("Apple","Orange","Mango")
print("Tuple before updation:",fruit_tuple)

fruit_list=list(fruit_tuple)       #convert tuple into list

fruit_list[1]="banana"             #update orange with banana

fruit_tuple=tuple(fruit_list)      #convert list into tuple
print("Tuple after updation:",fruit_tuple)
```

```python
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)

print(x)
```

```python
#Add Items
thistuple = ("apple", "banana", "cherry")

y = list(thistuple)
y.append("orange")
thistuple = tuple(y)
print(thistuple)
```

```python
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.remove("apple")
thistuple = tuple(y)
print(thistuple)
```

```python
#Unpack Tuples
#Unpacking a tuple:

fruits = ("apple", "banana", "cherry")
(green, yellow, red) = fruits

print(green)
print(yellow)
print(red)
```

```python
#Assign the rest of the values as a list called "red":

fruits = ("apple", "banana", "cherry", "strawberry", "raspberry","strawberry")

(green,yellow,*red) = fruits

print(green)
print(yellow)
print(red)
```

```python
fruits = ("apple", "mango", "papaya", "pineapple", "cherry","banana")

(green, *tropic,red) = fruits

print(green)
print(tropic)
print(red)
```

```python
#Join Tuples
tuple1 = (5,6,7,8)
tuple2 = (1, 2, 3)

tuple3 = tuple1 + tuple2
print(tuple3)
```

```python
#Multiply the fruits tuple by 3:

fruits = ("apple", "banana", "cherry")

mytuple = fruits * 3

print(mytuple)
```

```python
numbers = (10, 20, 50, 20, 30, 40)

print(len(numbers))        # 6
print(max(numbers))        # 50
print(min(numbers))        # 10
print(sum(numbers))        # 170
print(sorted(numbers))     # [10, 20, 20, 30, 40, 50]
print(numbers.count(20))   # 2
print(numbers.index(30))   # 2
```
